### Table of Income Statement :

An income statement is a financial statement that shows you the company’s income and expenditures. It also shows whether a company is making profit or loss for a given period. The income statement, along with balance sheet and cash flow statement, helps you understand the financial health of your business.
The income statement is also known as a profit and loss statement, statement of operation, statement of financial result or income, or earnings statement.

```python
Income_Statement = pd.read_csv('Income_Statement.csv')
Income_Statement.drop(Income_Statement.columns[Income_Statement.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
Income_Statement.head(8)
```

```python
Income_Statement.rename(columns={'ttm':'12/31/2020'},inplace=True)
Income_Statement= Income_Statement.transpose()
Income_Statement.reset_index(inplace=True)
Income_Statement.columns = Income_Statement.iloc[0]
Income_Statement.drop(Income_Statement.index[0],inplace=True)
Income_Statement.reset_index(inplace=True)
Income_Statement.drop('index',axis=1,inplace=True)
Income_Statement.rename(columns={'Breakdown':'Annual'},inplace=True)
```

```python
Income_Statement
```

```python
Income_Statement= Income_Statement.replace ('-',np.nan)
Income_Statement= Income_Statement.replace (',','',regex=True)
Income_Statement
```

```python
Income_Statement.columns
```

```python
Income_Statement.info()
```

```python
Income_Statement['Annual'] = pd.to_datetime(Income_Statement['Annual'].str.strip(), format="%m/%d/%Y" )
Income_Statement['Annual']
```

```python
cols =Income_Statement.columns
Income_Statement[cols[1:]] =Income_Statement[cols[1:]].astype(np.float64)
```

```python
Income_Statement.dtypes
```

```python
Income_Statement.isna().sum()
```

```python
Income_Statement.fillna(Income_Statement.mean(), inplace=True)
Income_Statement
```

```python
Income_Statement.describe()
```
